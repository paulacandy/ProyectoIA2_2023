```python
b = np.load('recycle_data_shuffled.npz')
print(b.files)
x_train=b['x_train']
x_test=b['x_test']
y_train=b['y_train']
y_test=b['y_test']
type(x_test)
print(np.shape(x_train),np.shape(x_test),np.shape(y_train),np.shape(y_test))
```

```python
#Primero con np.unique supe que sus valores venian de 0-255
#como vienen de 0-255 entonces:
x_train=x_train/255
x_test=x_test/255

#acá averiguo de qué tamaño viene la imagen
print(np.shape(x_train[0]))
x_train = resize(x_train, (np.shape(x_train)[0], 32, 32, 3))
x_test = resize(x_test, (np.shape(x_test)[0], 32, 32, 3))
```

```python
print(np.shape(x_train),np.shape(x_test),np.shape(y_train))
```

```python
print(np.unique(y_train))
```

```python
#Concateno x y y
x=np.concatenate((x_train, x_test))
y=np.concatenate((y_train, y_test))

y1=np.repeat(1, len(y))

np.save("x1", x)
np.save("y1", y1)
```
