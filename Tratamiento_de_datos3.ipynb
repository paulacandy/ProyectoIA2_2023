```python
ruta_1 = "/content/drive/MyDrive/IA2_PROYECTOFINAL/Dataset/3/Surgical waste dataset"
```

```python
NR = []

im_size = 32

for im in os.listdir(ruta_1):
  im = cv2.imread(os.path.join(ruta_1,im))

  img = cv2.resize(im, (im_size,im_size))
  NR.append(img)


print(len(NR))
```

```python
x=NR
y=np.repeat(0, len(NR))
```
